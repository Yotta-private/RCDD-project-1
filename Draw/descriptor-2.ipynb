```python
import seaborn as sns
import pandas as pd
import sys
import matplotlib.pyplot as plt
%matplotlib inline
#name :Name	ALogP	AMR	nAtom	nHeavyAtom	fragC	nHBAcc	nHBDon	nRing	nRotB	TopoPSA
df_gen = pd.read_table( 'generate_scls.csv',encoding='gbk')
df_train = pd.read_table( 'train_scls.csv',encoding='gbk')
print(df_gen.shape)
# df=df1.join(df2.np)
# sns.distplot( df[df.cat == 'DIV'].np)
# sns.distplot( df[df.cat == 'NAT'].np)
print(df_gen[:5])
print(df_train[:5])
df_gen=df_gen[df_gen['COMPLEXITY'] > 0.5]
df_train=df_train[df_train['COMPLEXITY'] > 0.5]
print(df_gen.shape)
# df_droptrain = df_train.COMPLEXITY.dropna(axis = 0)
```

```python
sns.set(style="ticks", palette="muted", color_codes=True)
sns.distplot(df_gen['COMPLEXITY'])
sns.distplot(df_train['COMPLEXITY'],color="r")
#sns.distplot(arrs)
# sum((df_gen.shape[0] - df_gen.count()) > 0)
```

```python

plt.figure(figsize=(15, 12))
plt.scatter(y=df_train.CYCLICITY[40000:50000], x=df_train.COMPLEXITY[40000:50000],alpha=0.9, color = 'g',label='Training set')
plt.scatter(y=df_gen.CYCLICITY[40000:50000], x=df_gen.COMPLEXITY[40000:50000],alpha=0.75,color="r", label='Generated set')
plt.xlim(10,100)
plt.ylim(30,100)
plt.legend(loc='best',fontsize=20)
plt.tick_params(labelsize='medium', width=3)
plt.tick_params(labelsize=20)
plt.xlabel('Complexity',fontsize=20)

plt.ylabel('Cyclicity',fontsize=20)  
plt.savefig('Scluster',dpi = 220)
plt.show()
```
