```python
from rdkit import DataStructs

fps = [AllChem.GetMorganFingerprintAsBitVect(mol,2,nBits=1024) for mol in genlist[:30000]]
```

```python
smililarity = DataStructs.DiceSimilarity(fps[0],fps[0])
smililarity 
```

```python
i = 0
j = 0
res = []
rest = []
for i, genmol in enumerate(fps):
    for j, trainmol in enumerate(fps):
        smililarity = DataStructs.DiceSimilarity(genmol,trainmol)
        if smililarity != 1:
            res.append(smililarity)
    res.sort()
    if i%500 ==0 : print(i)
    rest.append(res)
    res = []
```
