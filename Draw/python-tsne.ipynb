```python
Graph = pd.read_csv('withouttran.csv')
```

```python
Graph = Graph.dropna(axis=0,how='any')

Graph.head(5)


Graph = Graph.reset_index(drop=True)
```

```python
Graph.iloc[:,1:]
```

```python
G_embed = TSNE(n_components=2).fit_transform(Graph.iloc[:,1:])
```

```python
from sklearn.preprocessing import MinMaxScaler
G_embed_change = MinMaxScaler().fit_transform(G_embed)
```

```python
G_embed_change
```

```python
df = pd.DataFrame(np.transpose((G_embed_change[:,0],G_embed_change[:,1])))
df.colunms = ['x','y']
df.head(5)
```
